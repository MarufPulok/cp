```python
# why functions are useful
num = 5
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")

def is_even(number):
    return number % 2 == 0

if is_even(num):
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
# variable length arguments
def sum_all(*args):
    total = 0
    for number in args:
        total += number
    return total

print(sum_all(1, 2, 3, 4, 5))  # Outputs: 15
print(sum_all(10, 20))        # Outputs: 30
```

```python
# keyword arguments
def person_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

person_info(name="Alice", age=30, city="New York")
```

```python
def foo(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

foo(1, 2, 3, name="Bob", age=25)
foo()  # No arguments
foo(42)  # Only positional argument
foo(city="Los Angeles")  # Only keyword argument
```

```python
# return statements
def add(a, b):
    return a + b
result = add(3, 5)
print(f"The sum is: {result}")
```
