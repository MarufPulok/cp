```python
# example1: manage a todo list
todo_list = ["buy groceries", "clean the house", "pay bills"]

# adding a new task
todo_list.append("walk the dog")
print("Updated todo list:", todo_list)
# removing a completed task
todo_list.remove("pay bills")
print("Todo list after removing a task:", todo_list)

if "clean the house" in todo_list:
    print("Don't forget to clean the house today!")
```

```python
# example2: organizing student grades
grades = [85, 92, 78, 90, 88, 76, 95]
average_grade = sum(grades) / len(grades)
print("Average grade:", average_grade)

# finding highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print("Highest grade:", highest_grade)
print("Lowest grade:", lowest_grade)
```

```python
# example3: managing an inventory
inventory = ["sword", "shield", "potion"]
inventory.append("armor")
print("Current inventory:", inventory)
inventory.remove("potion")
print("Inventory after using a potion:", inventory)
if "sword" in inventory:
    print("You are ready for battle!")
else:
    print("You need a weapon!")
```

```python
# example4: collecting user feedbacks
feedbacks = []
feedbacks.append("Great service!")
feedbacks.append("Could improve the delivery time.")
print("User feedbacks:", feedbacks)
if len(feedbacks) > 0:
    print("Thank you for your feedback!")
else:
    print("No feedback received yet.")
```
