```python
# create a set
my_set = {1, 2, 3, 4, 5}
print(my_set, type(my_set))
```

```python
my_empty_set = set()
print(my_empty_set, type(my_empty_set))
```

```python
my_empty_set = set([1, 2, 3])
print(my_empty_set, type(my_empty_set))

list1 = [1, 2, 2, 3, 4, 4, 5]
unique_set = set(list1)
print(unique_set)
```

```python
# basic set operations
my_set.add(6)
print("After adding 6:", my_set)
my_set.add(6)  # adding duplicate
print("After adding 6 again:", my_set)
```

```python
my_set.remove(3)
print("After removing 3:", my_set)
my_set.remove(10)  # removing non-existing element raises KeyError
print("After trying to remove 10:", my_set)
```

```python
my_set.discard(18)  # removing non-existing element does not raise error
print("After discarding 18:", my_set)
```

```python
my_set.discard(4)
print("After discarding 4:", my_set)
```

```python
# pop method
popped_element = my_set.pop()
print("Popped element:", popped_element) # first in first out (arbitrary element)
print("Set after popping an element:", my_set)
popped_element = my_set.pop()
print("Set after popping another element:", my_set)
```

```python
# set membership
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)  # True
print(10 in my_set)  # False
```

```python
# mathematical operation
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# union
union_set = set1.union(set2)
print("Union:", union_set)
# intersection
intersection_set = set1.intersection(set2)
print("Intersection:", intersection_set)
# difference
difference_set = set1.difference(set2)
print("Difference (set1 - set2):", difference_set)
# symmetric difference
symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric Difference:", symmetric_difference_set)
# subset and superset
print("Is set1 a subset of union_set?", set1.issubset(union_set))
print("Is union_set a superset of set1?", union_set.issuperset(set1))

# intersection update
set1.intersection_update(set2)
print("Set1 after intersection update with set2:", set1)
```

```python
# difference update
set1 = {1, 2, 3, 4}  # reset set1
set2 = {3, 4, 5, 6}  # reset set2
set1.difference_update(set2)
print("Set1 after difference update with set2:", set1)
```

```python
# set methods
set1 = {1, 2, 3}
set2 = {4, 5, 6}

print("Is set1 a subset of set2?", set1.issubset(set2))  # False
print("Is set1 a superset of set2?", set1.issuperset(set2))  # False
```

```python
lst = [1, 2, 3, 4, 5, 5, 6, 6]
unique_elements = set(lst)
print("Unique elements from the list:", unique_elements)
```

```python
# counting unique words in text
text = "In this tutorial we will learn about sets in Python. Sets are useful for storing unique items."
words = text.split()
unique_words = set(words)
print("Unique words in the text:", unique_words)
print("Number of unique words:", len(unique_words))
```
